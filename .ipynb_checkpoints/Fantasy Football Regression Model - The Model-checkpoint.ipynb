Now, as mentioned in my EDA notebook, I want focus on trying to maximize offsensive scoring in fantasy football, and I want to go about this by finding out which teams defenses are best at _preventing_ other teams from scoring offensive fantasy football points. For reference, [ESPN's Standard Scoring System](http://www.espn.com/fantasy/football/ffl/story?page=fflrulesstandardscoring) for offensive player is as follows:

> **Quarterbacks (QB), Running Backs (RB), Wide Receivers (WR), Tight Ends (TE)**  
- 6 pts per rushing or receiving TD
- 6 pts for player returning kick/punt for TD
- 6 pts for player returning or recovering a fumble for TD
- 4 pts per passing TD
- 2 pts per rushing or receiving 2 pt conversion (note: teams do not receive points for yardage gained during the conversion)
- 2 pts per passing 2 pt conversion
- 1 pt per 10 yards rushing or receiving
- 1 pt per 25 yards passing  
*Bonus Points* (note: the player must score a touchdown to score the points)  
- 2 pts per rushing or receiving TD of 40 yards or more
- 2 pts per passing TD of 40 yards or more  
*Penalty Points*    
- -2 pts per intercepted pass
- -2 pts per fumble lost  

With this in mind, I want to find defensive teams that produce the most interceptions and lost fumbles while preventing TDs and yards.
```python
# %R is the 'magic' to use in order to have the subsequent line of code processed by the R kernel
%R def_gen = read.csv('data/pff-data/scaled_data/def_gen_scaled.csv', header=TRUE, sep=",")
%R def_pass = read.csv('data/pff-data/scaled_data/def_pass_scaled.csv', header=TRUE, sep=",")
%R def_rush = read.csv('data/pff-data/scaled_data/def_rush_scaled.csv', header=TRUE, sep=",")
%R cb = read.csv('data/pff-data/scaled_data/cb_scaled.csv', header=TRUE, sep=",")
%R di = read.csv('data/pff-data/scaled_data/di_scaled.csv', header=TRUE, sep=",")
%R edge = read.csv('data/pff-data/scaled_data/edge_scaled.csv', header=TRUE, sep=",")
%R lb = read.csv('data/pff-data/scaled_data/lb_scaled.csv', header=TRUE, sep=",")
%R s = read.csv('data/pff-data/scaled_data/s_scaled.csv', header=TRUE, sep=",")
%R pbp = read.csv(file="data/nflscrapR-data/scaled_data/pbp_scaled.csv", header=TRUE, sep=",")
%R games = read.csv(file="data/nflscrapR-data/scaled_data/games_scaled.csv", header=TRUE, sep=",")
```

Awesome, now that all of the data is read into the R kernel, I can begin building my models. I will build my first model using the data from the `games` dataset. For this model, I will use `home_score_scaled` and `away_score_scaled` as my DVs and `home_team` and `away_team` as my predictor variables. Since the values of `home_score` and `away_score` use the same units, I don't really _need_ to use the scaled value; however, since I plan to amalgamate the coefficients of this model with the coefficients of models using variables that have different units, I will stick with the scaled values, otherwise I might introduce some bias into my overall predictions.

```python
%R model = lm(cbind(home_score_scaled, away_score_scaled) ~ home_team + away_team, data = games)
```

Great, with that taken care of, we might as well take a look at the summary of the model. I want to see how each IV value relates to each DV, so I will use `summary.aov` instead of just `summary` as `summary.aov` will output details in relation to each DV:

```python
%R print(summary.aov(model))
```

Alright, so it looks like there is significance in regards to how both IVs relate to both DVs! Now the next thing I want to do with this model is collect just the coefficients as these are the values I really care about from the model since they essentially represent how the teams 'stack up' to each other. So, I will grab the coefficients from the model and then I'll cast these into a dataframe, and then I will 'transfer' that dataframe from the R kernel over to the Python kernel:

```python
%R coefs = model$coefficients
%R df = as.data.frame(coefs, keep.rownames=TRUE)
scores_by_team_model = %R df
scores_by_team_model.head()
```

Since I will only be using overall scores and not pass rush, run def, and coverage scores, and all of my positional groups datasets contain the overall score column, I am going to extract just the team and overall score column from each dataset and create a single dataset so that I can create one model instead of having one for each positional group. Something that I'm realizing might have a biasing effect on my model from these datasets is the fact that not all players really end up playing all that often (which is captured by the `TOTAL_SNAPS` column), and according to the correlations from the EDA dataset this directly links to having a lower overall score, which would likely have a negative impact on my model. To address this, I will drop rows from the positional groups datasets in which the `TOTAL_SNAPS` is less than the average number of `TOTAL_SNAPS` from that dataset before grabbing the `TEAM` and `OVERALL_scaled` columns from the dataset and appending it to my new dataset. Lets do all of this now:

```python
# Drop rows where total snaps is less than the mean for that dataset
cb = cb[cb['TOTAL_SNAPS'] > cb['TOTAL_SNAPS'].mean()]
edge = edge[edge['TOTAL_SNAPS'] > edge['TOTAL_SNAPS'].mean()]
s = s[s['TOTAL_SNAPS'] > s['TOTAL_SNAPS'].mean()]
di = di[di['TOTAL_SNAPS'] > di['TOTAL_SNAPS'].mean()]
lb = lb[lb['TOTAL_SNAPS'] > lb['TOTAL_SNAPS'].mean()]
```

I just realized that this is going to cause a significant problem: I plan to use the `OVERALL_scaled` column for my model; however, these values are scaled based on the entire, individual, datasets, and I just significantly reduced the size of these datasets (and likely removed many of the 'lower teir' players on the lower end of the scale) thereby completely screwing up the scaling. Not only that, but when I put these together like I'm about to, this will also significantly alter the dataset and how scaling should be done with it. Because of this, I need to capture the original `OVERALL` column from each dataset, and then after putting all of these together into a single, larger dataset I will need to redo the scaling. I'm glad I caught this! Let's see what we can do:

```python
# create singular positional groups dataset
pos_groups_data = cb[['TEAM', 'OVERALL']]
pos_groups_data = pos_groups_data.append(edge[['TEAM', 'OVERALL']], ignore_index = True)
pos_groups_data = pos_groups_data.append(s[['TEAM', 'OVERALL']], ignore_index = True)
pos_groups_data = pos_groups_data.append(di[['TEAM', 'OVERALL']], ignore_index = True)
pos_groups_data = pos_groups_data.append(lb[['TEAM', 'OVERALL']], ignore_index = True)

# Reset the index in case strange indexes could cause issues in the future
pos_groups_data.reset_index(drop = True, inplace = True)
pos_groups_data.head()
```

Now that I have the dataset, I need to take care of the scaling. I will use the same approach to what I did in my EDA notebook. In fact, I will just copy over the EDA I created in that notebook and apply it here:

```python
from sklearn import preprocessing

def scale_cols(df, cols, dropna = True):
    df_copy = df.copy()  # using a copy so changes aren't applied in place for debugging purposes
    if dropna:
        # NA values cause problems with scaling and StandardScaler has no way of just ignoring these values
        # so drop rows with NA values before scaling. They are few and far between so shouldn't be an issue
        df_copy.dropna(axis = 0, subset = cols,inplace = True)
        # have to reset the index otherwise later combining becomes an issue
        df_copy.reset_index(drop = True, inplace = True)
    temp = df_copy[cols]
    cols = [col + "_scaled" for col in cols]
    scaler = preprocessing.StandardScaler()
    scaled_df = scaler.fit_transform(temp)
    scaled_df = pd.DataFrame(scaled_df, columns=cols)
    df_copy[cols] = scaled_df
    return df_copy

pos_groups_data = scale_cols(pos_groups_data, ["OVERALL"])
pos_groups_data = pos_groups_data[["TEAM", "OVERALL_scaled"]]
pos_groups_data.head()
```

Alright, now that everything is in place I need to start deciding what will go into my final team scores and what I'll leave out. As mentioned previously, I want to build a model based on defensive team strength, so what really matters in this context are models that were built using the defensive team as the basis, which doesn't apply to all of the models I built. On the other hand, I will include the following model coefficients in my final team scores as all of these are related to defensive team strength, either by using `defteam` as the predictor variable or by coming from defensive positional group stats, thus telling me about defensive team strength:

- yards_by_defteam_model
- interceptions_model
- fumbles_model

Now that all of this is sorted out, I want to get all of these coefficients into a single dataset for ease of use. To do this, I will use the Pandas `merge()` method to join my datasets using essentially an SQL inner join:

```python
coefs = yards_by_defteam_model.merge(interceptions_model, on = 'team')
coefs = coefs.merge(fumbles_model, on = 'team')
coefs = coefs.merge(TDs_model, on = 'team')
coefs = coefs.merge(pos_groups_model, on = 'team')
coefs
```

Looking good! One issue I'm noticing, however, is that the columns in which logistic regression was used (`interceptions_coef`, `fumbles_coef`, and `TDs_coef`) all have a much larger values compared to the other columns, likely because of the difference in the way logistic regression is calculated versus standard linear regression. Because of this, I will scale these columns using the UDF from above in order to get all columns on a level playing field. Then, I can apply my own weighting so that each column is proportionate to the number of fantasy football points that they represent:

```python
coefs_scaled = scale_cols(coefs, cols = list(coefs.columns)[1:])
coefs_scaled
```

Perfect! Now I was originally planning to weight yards, touchdowns, interceptions, and fumbles according to the amount of points they actually contribute to fantasy football scores; however, this introduces some issues because touchdowns are by far worth the most points (i.e. 6) while yards are worth relatively few when you boil it down to the yard (i.e. 0.1 points per rushing or receiving yard or 0.04 points per passing yard), yet yards are a much more common means of receiving fantasy points than touchdowns. I'm not exactly certain how to go about weighting these, and I am tempted to simply leave them unweighted, but another issue is that I have three columns that are related to yards but just one related to touchdowns, thus not weighting yards might bias the model toward teams that are good at stopping yardage. Of course, yards are a more frequent occurance than touchdowns, so honestly this is a good thing as [this CBS Boston article](https://boston.cbslocal.com/2017/09/19/nfl-scoring-touchdown-problem/) explains that there is an average of just under 2.5 touchdowns per team per game. For this reason, I won't weight the TDs column since TDs are worth a significant number of points but I also won't weight the yards columns as this will help to factor in the more likely contribution of yards to fantasy football points.

Now I still have to consider what to do with interceptions and fumbles. These are worth negative points, so at the very least I will weight these columns by -1; however, these events, like touchdowns, don't happen terribly often. In fact, according to this [NFL by the Numbers](http://www.sportsonearth.com/article/64441086/nfl-statistical-analysis-average-nfl-game) article, they are essentially just as common as touchdowns, yet their worth -1/3 times the points compared to touchdowns. For this reason, I will weight the `interceptions_coef_scaled` and `fumbles_coef_scaled` columns by -0.34.

Finally, I must consider how to weigh the `overall_score_coef_scaled` column. Unfortunately, this column doesn't directly translate into points but rather is an indicator of the strength of a defensive team overall. Since this represents the overall strength of a defense, the higher the number the more difficult, at least theoretically, it would be to score fantasy points against that team, thus this column should be weighted by some negative number at the very least. Beyond knowing that I should weight this by a negative number, though, I really don't know how to quantify this overall defensive team strength in relation to ability to stop yards/touchdowns or cause fumbles/interceptions. For this reason, I will just weight this column by -1. 

I realize that it is certainly not ideal for me to come up with this weighting myself, but it's the best option I have right now and I will be sure to see how this season goes and take detailed notes throughout so that I can iterate back over my model to try to improve it next year. For now, let's take care of the weighting:

```python
coefs_weighted = coefs_scaled[['team', 'yards_gained_coef_scaled', 'air_yards_coef_scaled', 
                               'yards_after_catch_coef_scaled', 'TDs_coef_scaled']].copy()
coefs_weighted['interceptions_coef_weighted'] = coefs_scaled['interceptions_coef_scaled'] * -0.34
coefs_weighted['fumbles_coef_weighted'] = coefs_scaled['fumbles_coef_scaled'] * -0.34
coefs_weighted['overall_score_coef_weighted'] = coefs_scaled['overall_score_coef_scaled'] * -1
coefs_weighted
```

Great! Now I just need to add up the values in each row in order to obtain a score for each team. This score will represent the strength of a teams defense, with a more negative value representing an overall more difficult team to score fantasy points against and a more positive value meaning just the opposite:

```python
team_scores = pd.DataFrame(coefs_weighted['team'])
team_scores['score'] = coefs_weighted.iloc[:,1:].sum(axis = 1)
team_scores
```

Awesome!
```python
team_scores.loc[32,:] = ['BYE', 0]
team_scores
```
