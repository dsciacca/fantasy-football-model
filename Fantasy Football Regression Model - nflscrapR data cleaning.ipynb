The ones I won't drop because I believe they _could_ have a significant effect on the model (EDA will reveal whether or not this is the case) include: `goal_to_go`, `first_down_rush`, `first_down_pass`, `incomplete_pass`, `interception`, `safety`, `tackled_for_loss`, `fumble_lost`, `qb_hit`, `rush_attempt`, `pass_attempt`, `sack`, `touchdown`, `pass_touchdown`, `rush_touchdown`,` return_touchdown`, `fumble`, and `complete_pass`.

```python
keepers = ["goal_to_go", "first_down_rush", "first_down_pass", "incomplete_pass", "interception", 
           "safety", "tackled_for_loss", "fumble_lost", "qb_hit", "rush_attempt", "pass_attempt", 
           "sack", "touchdown", "pass_touchdown", "rush_touchdown"," return_touchdown", "fumble", 
           "complete_pass"]
print("Length of keepers: " + str(len(keepers)))
to_drop = []
for col in reg_pbp18.columns:
    if col in keepers or str(reg_pbp18[col].dtype) == 'object':
        continue
    temp = list(reg_pbp18[col].unique())
    temp = [x for x in temp if str(x) != 'nan']
    temp.sort()
    if temp == [0, 1] or temp == [0] or temp == [1]:
        to_drop.append(col)
print("Length of to_drop: " + str(len(to_drop)))
```

Great! Now let's drop all of the `to_drop` columns from each of our datasets. These only apply to the play-by-play datasets, so let's make sure to only do this to these columns to save a little bit of computational power:

```python
for key, value in data_dict.items():
    if 'pbp' in key:
        value.drop(to_drop, axis = 1, inplace = True, errors = 'ignore') # some datasets don't contain touchback col so need to supress errors
reg_pbp18.shape
```

Because of this oddity, and due to the fact that not every team is present in these datasets which could lead to bias in my future models, I will not use these `post_pbp` datasets going forward, and since this is also likely the case in the `post_games` datasets, I will also not use these going forward either.

Something I also noticed in looking at this is that the Chargers are listed as `SD` in some of these datasets, but listed as `LAC` in others (see output for post_pbp18 vs post_pbp13, for example). This is because the Chargers moved from San Diego to LA a few years back, resulting in a change to their team name. Similarly, the Rams moved from St Louis to LA a few years back, so their team abbreviation changed from `STL` to `LAR`, or, as seen in the output above, `LA`. Additionally, in my past experience with this kind of data, I know that sometimes the Jacksonville Jaguares are listed as `JAC` and others listed as `JAX`. I should take care of these discrepencies now:

```python
for key, value in data_dict.items():
    value['home_team'] = value['home_team'].str.replace("JAC", "JAX")
    value['away_team'] = value['away_team'].str.replace("JAC", "JAX")
    value['home_team'] = value['home_team'].str.replace("SD", "LAC")
    value['away_team'] = value['away_team'].str.replace("SD", "LAC")
    value['home_team'] = value['home_team'].str.replace("STL", "LA")
    value['away_team'] = value['away_team'].str.replace("STL", "LA")
    if 'pbp' in key:
        value['posteam'] = value['posteam'].str.replace("JAC", "JAX")
        value['defteam'] = value['defteam'].str.replace("JAC", "JAX")
        value['posteam'] = value['posteam'].str.replace("SD", "LAC")
        value['defteam'] = value['defteam'].str.replace("SD", "LAC")
        value['posteam'] = value['posteam'].str.replace("STL", "LA")
        value['defteam'] = value['defteam'].str.replace("STL", "LA")
```

Another issue I have with these column names is that many of them have a weird combination of snake case and camel case, making them difficult to read and even more difficult for keeping straight when trying to reference these columns. Let's turn this into strictly snake case with each word or acronym being all lower case, which is more like the pbp datasets. The function I originally planned to use that I obtained from StackOverflow resulted in issues with columns with `TDs` present. This is not consistent with the pbp datasets anyway since this dataset instead uses touchdown(s), so while I'm at it I'll have the UDF convert `TDs` to `touchdowns` and `TD` to `touchdown`:

NOTE to self for later: RACR stands for Receiver Air Conversion Ratio and PACR stands for Passing Air Conversion Ratio

```python
# Code adapted from: https://stackoverflow.com/questions/1175208/elegant-python-function-to-convert-camelcase-to-snake-case
first_cap_re = re.compile('(.)([A-Z][a-z]+)')
all_cap_re = re.compile('([a-z0-9])([A-Z])')
def convert(name):
    name = name.replace("TDs", "touchdowns")
    name = name.replace("TD", "touchdown")
    s1 = first_cap_re.sub(r'\1_\2', name)
    s1 = all_cap_re.sub(r'\1_\2', s1).lower()
    return s1.replace("__", "_")
```

```python
game_player_receiving.columns = game_player_receiving.columns.to_series().apply(convert)
game_player_rushing.columns = game_player_rushing.columns.to_series().apply(convert)
game_player_passing.columns = game_player_passing.columns.to_series().apply(convert)
with pd.option_context("display.max_rows", 10, "display.max_columns", None):
    print("Receiving:")
    display(game_player_receiving)
    print("Rushing:")
    display(game_player_rushing)
    print("Passing:")
    display(game_player_passing)
```
